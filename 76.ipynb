```python
import matplotlib.pyplot as plt 
try:
    data_file = open("opendata.stat.txt")
    data = data_file.read()
    counter = 0
    size_pension = []
    date = []
    for i in data.split("\n"):
        if i.find("Средняя пенсия")!=-1 and i.find("2018")!=-1 and i.find("Забайкальский край")!=-1:
            size_pension.append(int(i.split(",")[3]))
            date.append(i.split(",")[2])
    plt.figure(figsize=(12, 6))
    plt.plot(date, size_pension)
    plt.show()
    print(f"средняя пенсия в забайкальском крае составляет: {(sum(size_pension)/len(size_pension)): .2f} р.")
except Exception as ex:
    print("при работе возникла ошибка: ", ex)


```

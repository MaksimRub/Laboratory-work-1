```python
import requests
class myFile:
    file = None
    mode = ""
    def __init__(self, file_name, mode):
        try:
            if mode == "read":
                self.file = open (file_name, "r")
            elif mode == "write":
                self.file = open (file_name, "w")
            elif mode == "append":
                self.file = open (file_name, "a")
            elif mode == "url":
                self.file = requests.get(file_name)
            else:
                print("неверный режим открытия файла")
        except Exception as ex:
            print("ошибка при работе с файлом: ", ex)
    def write(self,text):
        try:
            self.file.write(text)
        except Exception as ex:
            print("при записи в файл возникла ошибка: ",ex)
    def read(self):
        try:
            print(self.file.read())
        except Exception as ex:
            print("при чтении из файла возникла ошибка: ",ex)
        
```

```python

try:
    data_file = open("opendata.stat.txt")
    data = data_file.read()
    counter = 0
    sum1 = 0
    for i in data.split("\n"):
        if i.find("Средняя пенсия")!=-1 and i.find("2018")!=-1 and i.find("Забайкальский край")!=-1:
            sum1 += int(i.split(",")[3])
            counter+=1
    print(sum1/counter)
except (ZeroDivisionError):
    print("деление на 0")
except:
    print("ошибка работы с файлом")


```
